### Add additional features

```python
indicators_df = unnested_df.copy()
indicators_df = indicators_df[
    indicators_df["entry_order_placement_time"] >= "2025-08-20T09:50:00.000000-04:00"
]
indicators_df["di_slope_plus"] = indicators_df["di_slopes"].apply(lambda x: x[0])
indicators_df["di_slope_minus"] = indicators_df["di_slopes"].apply(lambda x: x[1])
indicators_df.drop(columns=["di_slopes"], inplace=True)
indicators_df["adx_to_di_plus_ratio"] = (
    indicators_df["adx_value"] / indicators_df["di_plus"]
)
indicators_df["adx_to_di_minus_ratio"] = (
    indicators_df["adx_value"] / indicators_df["di_minus"]
)
indicators_df["di_ratio"] = indicators_df["di_plus"] / indicators_df["di_minus"]
indicators_df["di_difference"] = indicators_df["di_plus"] - indicators_df["di_minus"]
indicators_df["volume_ratio"] = (
    indicators_df["volume_value"] / indicators_df["volume_avg"]
)
indicators_df["atr_ratio"] = indicators_df["atr_value"] / indicators_df["atr_avg"]
indicators_df.drop(columns=["atr_avg", "atr_value"], inplace=True)
numeric_indicators_df = indicators_df.apply(pd.to_numeric, errors="coerce")
```

```python
indicators_df["date"] = (
    indicators_df["entry_order_placement_time"].str.split("T").str[0]
)
indicators_df["day_of_the_week"] = indicators_df["date"].apply(
    lambda x: pd.Timestamp(x).day_name()
)

indicators_df["hour"] = indicators_df["entry_order_placement_time"].apply(
    lambda x: int(x.split("T")[1].split(":")[0])
)
```

```python
numeric_cols = [
    "adx_slope",
    "adx_to_di_minus_ratio",
    "adx_to_di_plus_ratio",
    "adx_value",
    "atr_ratio",
    "atr_slope",
    "atr_stop_loss_to_price_ratio",
    "di_minus",
    "di_plus",
    "di_difference",
    "di_ratio",
    "di_slope_minus",
    "di_slope_plus",
    "net_pnl_usd",
    "rsi_mean",
    "rsi_ratio",
    "rsi_trend_reverse_slope_long",
    "rsi_trend_reverse_slope_value",
    "rsi_value",
    "volume_ratio",
    "volume_slope",
]

MIN_COUNT = 0

threshold_results = []

savings_ratio_threshold = -0.001
filter_ratio_threshold = 0.1
for col in numeric_cols:
    if col == "net_pnl_usd":
        continue

    values = indicators_df[col].dropna().unique()
    if values.size == 0:
        continue
    sorted_vals = np.sort(values)

    found_threshold_lower = None
    filterd_lower_count = 0
    ratio_trades_lower = 0
    ratio_positive_negative_lower = 0

    found_threshold_higher = None
    filterd_higher_count = 0
    ratio_trades_higher = 0
    ratio_positive_negative_higher = 0

    total_pnl = indicators_df["net_pnl_usd"].sum()
    total_trades = len(indicators_df)
    for thresh in sorted_vals[::-1]:
        subset_le = indicators_df[indicators_df[col] <= thresh]
        filterd_lower_count = len(subset_le)
        is_filtered_only_negative = (subset_le["net_pnl_usd"] < 0).all()
        positive_pnl_sum = (
            subset_le[subset_le["net_pnl_usd"] > 0]["net_pnl_usd"].sum()
            if len(subset_le) > 0
            else 0
        )
        negative_pnl_sum = (
            subset_le[subset_le["net_pnl_usd"] < 0]["net_pnl_usd"].sum()
            if len(subset_le) > 0
            else 0
        )
        ratio_positive_negative_lower = (
            (positive_pnl_sum + negative_pnl_sum) / abs(total_pnl)
            if total_pnl != 0
            else 0
        )
        ratio_trades_lower = (
            filterd_lower_count / total_trades if total_trades > 0 else 0
        )
        if (
            len(subset_le) >= MIN_COUNT
            and ratio_positive_negative_lower <= savings_ratio_threshold
            and (
                ratio_trades_lower <= filter_ratio_threshold
                or (
                    ratio_trades_lower > filter_ratio_threshold
                    and is_filtered_only_negative
                )
            )
        ):
            found_threshold_lower = f"<= {round(float(thresh), 2)}"
            break

    for thresh in sorted_vals:
        subset_ge = indicators_df[indicators_df[col] >= thresh]
        filterd_higher_count = len(subset_ge)
        is_filtered_only_negative = (subset_ge["net_pnl_usd"] < 0).all()
        positive_pnl_sum = (
            subset_ge[subset_ge["net_pnl_usd"] > 0]["net_pnl_usd"].sum()
            if len(subset_ge) > 0
            else 0
        )
        negative_pnl_sum = (
            subset_ge[subset_ge["net_pnl_usd"] < 0]["net_pnl_usd"].sum()
            if len(subset_ge) > 0
            else 0
        )
        ratio_positive_negative_higher = (
            (positive_pnl_sum + negative_pnl_sum) / abs(total_pnl)
            if total_pnl != 0
            else 0
        )
        ratio_trades_higher = (
            filterd_higher_count / total_trades if total_trades > 0 else 0
        )
        if (
            len(subset_ge) >= MIN_COUNT
            and ratio_positive_negative_higher <= savings_ratio_threshold
            and (
                ratio_trades_higher <= filter_ratio_threshold
                or (
                    ratio_trades_higher > filter_ratio_threshold
                    and is_filtered_only_negative
                )
            )
        ):
            found_threshold_higher = f">= {round(float(thresh), 2)}"
            break

    threshold_results.append(
        {
            "column": col,
            "neg_pnl_lower_th": found_threshold_lower,
            "filt_lower_count": filterd_lower_count,
            "ratio_lower": round(ratio_trades_lower, 3),
            "save_lower_pct": round(ratio_positive_negative_lower * 100, 1),
            "neg_pnl_upper_th": found_threshold_higher,
            "filt_higher_count": filterd_higher_count,
            "ratio_higher": round(ratio_trades_higher, 3),
            "save_higher_pct": round(ratio_positive_negative_higher * 100, 1),
        }
    )

threshold_df = pd.DataFrame(threshold_results)
threshold_df.set_index("column", inplace=True)
threshold_df
```

```python
sum_filtered_pnl = indicators_df["net_pnl_usd"].sum()
high_total_positive_trades = len(indicators_df[indicators_df["net_pnl_usd"] >= 5])
low_total_positive_trades = len(
    indicators_df[
        (indicators_df["net_pnl_usd"] > 0) & (indicators_df["net_pnl_usd"] < 5)
    ]
)
total_negative_trades = len(indicators_df[indicators_df["net_pnl_usd"] < 0])
total_trades_filtered = len(indicators_df)
print(
    f"Filtered PnL: {sum_filtered_pnl}, Total Trades: {total_trades_filtered}, pnl per trade: {sum_filtered_pnl / total_trades_filtered if total_trades_filtered > 0 else 0}"
    f"\nPositive Trades: {high_total_positive_trades} + {low_total_positive_trades}, Negative Trades: {total_negative_trades}"
    f"\nPnL per Positive Trade (Potential): {indicators_df[indicators_df['net_pnl_usd'] > 0]['net_pnl_usd'].sum():.1f}"
)
```
