## Unnesting and additional metrics

```python
indicators_df["di_slope_plus"] = indicators_df["di_slopes"].apply(
    lambda x: float(
        str(
            x.split(
                "(",
            )[2]
        ).split(
            ")",
        )[0]
    )
)
indicators_df["di_slope_minus"] = indicators_df["di_slopes"].apply(
    lambda x: float(
        str(
            x.split(
                "(",
            )[3]
        ).split(
            ")",
        )[0]
    )
)
indicators_df.drop(columns=["di_slopes"], inplace=True)
indicators_df["adx_to_di_plus_ratio"] = (
    indicators_df["adx_value"] / indicators_df["di_plus"]
)
indicators_df["adx_to_di_minus_ratio"] = (
    indicators_df["adx_value"] / indicators_df["di_minus"]
)
indicators_df["di_ratio"] = indicators_df["di_plus"] / indicators_df["di_minus"]
indicators_df["di_difference"] = indicators_df["di_plus"] - indicators_df["di_minus"]
indicators_df["volume_ratio"] = (
    indicators_df["volume_value"] / indicators_df["volume_avg"]
)
indicators_df["atr_ratio"] = indicators_df["atr_value"] / indicators_df["atr_avg"]
indicators_df.drop(columns=["atr_avg", "atr_value"], inplace=True)
indicators_filtered_df = indicators_df[
    [
        "pnl_usd",
        "adx_slope",
        "adx_to_di_minus_ratio",
        "adx_to_di_plus_ratio",
        "adx_value",
        "atr_ratio",
        "atr_slope",
        "atr_stop_loss_to_price_ratio",
        "di_minus",
        "di_plus",
        "di_difference",
        "di_ratio",
        "di_slope_minus",
        "di_slope_plus",
        "rsi_mean",
        "rsi_ratio",
        "rsi_trend_reverse",
        "rsi_trend_reverse_slope_long",
        "rsi_trend_reverse_slope_value",
        "rsi_value",
        "volume_ratio",
        "volume_slope",
        "hour",
    ]
]
indicators_filtered_df
```

```python
MIN_COUNT = 1

numeric_cols = indicators_filtered_df.select_dtypes(include=["number"]).columns.tolist()

threshold_results = []

savings_ratio_threshold = (
    -0.001
)  # ratio of positive to negative PnL sums to consider a threshold valid
filter_ratio_threshold = (
    0.1  # ratio of trades to total trades to consider a threshold valid
)
for col in numeric_cols:
    if col == "pnl_usd":
        continue

    values = indicators_filtered_df[col].dropna().unique()
    if values.size == 0:
        continue
    sorted_vals = np.sort(values)

    found_threshold_lower = None
    filterd_lower_count = 0
    ratio_trades_lower = 0
    ratio_positive_negative_lower = 0

    found_threshold_higher = None
    filterd_higher_count = 0
    ratio_trades_higher = 0
    ratio_positive_negative_higher = 0

    total_pnl = indicators_filtered_df["pnl_usd"].sum()
    total_trades = len(indicators_filtered_df)
    for thresh in sorted_vals[::-1]:
        subset_le = indicators_filtered_df[indicators_filtered_df[col] <= thresh]
        filterd_lower_count = len(subset_le)
        is_filtered_only_negative = (subset_le["pnl_usd"] < 0).all()
        positive_pnl_sum = (
            subset_le[subset_le["pnl_usd"] > 0]["pnl_usd"].sum()
            if len(subset_le) > 0
            else 0
        )
        negative_pnl_sum = (
            subset_le[subset_le["pnl_usd"] < 0]["pnl_usd"].sum()
            if len(subset_le) > 0
            else 0
        )
        ratio_positive_negative_lower = (
            (positive_pnl_sum + negative_pnl_sum) / abs(total_pnl)
            if total_pnl != 0
            else 0
        )
        ratio_trades_lower = (
            filterd_lower_count / total_trades if total_trades > 0 else 0
        )
        if (
            len(subset_le) >= MIN_COUNT
            and ratio_positive_negative_lower <= savings_ratio_threshold
            and (
                ratio_trades_lower <= filter_ratio_threshold
                or (
                    ratio_trades_lower > filter_ratio_threshold
                    and is_filtered_only_negative
                )
            )
        ):
            found_threshold_lower = f"<= {round(float(thresh), 2)}"
            break

    for thresh in sorted_vals:
        subset_ge = indicators_filtered_df[indicators_filtered_df[col] >= thresh]
        filterd_higher_count = len(subset_ge)
        is_filtered_only_negative = (subset_ge["pnl_usd"] < 0).all()
        positive_pnl_sum = (
            subset_ge[subset_ge["pnl_usd"] > 0]["pnl_usd"].sum()
            if len(subset_ge) > 0
            else 0
        )
        negative_pnl_sum = (
            subset_ge[subset_ge["pnl_usd"] < 0]["pnl_usd"].sum()
            if len(subset_ge) > 0
            else 0
        )
        ratio_positive_negative_higher = (
            (positive_pnl_sum + negative_pnl_sum) / abs(total_pnl)
            if total_pnl != 0
            else 0
        )
        ratio_trades_higher = (
            filterd_higher_count / total_trades if total_trades > 0 else 0
        )
        if (
            len(subset_ge) >= MIN_COUNT
            and ratio_positive_negative_higher <= savings_ratio_threshold
            and (
                ratio_trades_higher <= filter_ratio_threshold
                or (
                    ratio_trades_higher > filter_ratio_threshold
                    and is_filtered_only_negative
                )
            )
        ):
            found_threshold_higher = f">= {round(float(thresh), 2)}"
            break

    threshold_results.append(
        {
            "column": col,
            "neg_pnl_lower_th": found_threshold_lower,
            "filt_lower_count": filterd_lower_count,
            "ratio_lower": round(ratio_trades_lower, 3),
            "save_lower_pct": round(ratio_positive_negative_lower * 100, 1),
            "neg_pnl_upper_th": found_threshold_higher,
            "filt_higher_count": filterd_higher_count,
            "ratio_higher": round(ratio_trades_higher, 3),
            "save_higher_pct": round(ratio_positive_negative_higher * 100, 1),
        }
    )

threshold_df = pd.DataFrame(threshold_results)
threshold_df.set_index("column", inplace=True)
threshold_df
```
